```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #   3D-

# 1.  
x1 = np.linspace(-5, 10, 500)
x2 = np.linspace(-5, 10, 500)
X1, X2 = np.meshgrid(x1, x2)
Y = np.cos(X1) / (1 + np.exp(-2 * X2))

#    graph.csv
df = pd.DataFrame({'x1': X1.flatten(), 'x2': X2.flatten(), 'y': Y.flatten()})
df.to_csv('graph.csv', index=False)

#     
fig = plt.figure(figsize=(15, 10))

#  2D : y(x1)   x2
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x1, Y[0, :], label='y(x1)   x2')
ax1.scatter(x1, Y[0, :], color='red')
ax1.set_title(' y  x1   x2')
ax1.set_xlabel('x1')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

#  2D : y(x2)   x1
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x2, Y[:, 0], label='y(x2)   x1')
ax2.scatter(x2, Y[:, 0], color='blue')
ax2.set_title(' y  x2   x1')
ax2.set_xlabel('x2')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid(True)

# 3D 
ax3 = fig.add_subplot(2, 1, 2, projection='3d')
ax3.plot_surface(X1, X2, Y, cmap='viridis')
ax3.set_title('3D   y = cos(x1)/(1 + exp(-2 * x2))')
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.set_zlabel('y')

#      
plt.tight_layout()
plt.show()
```
